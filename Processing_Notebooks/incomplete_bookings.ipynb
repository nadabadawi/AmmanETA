### Jan Incomplete Bookings

```python
import pandas as pd

def process_files(file_prefix, start_day, end_day):
    for day in range(start_day, end_day + 1):
        file_name = f"{file_prefix}{day:02d}.csv"
        print(f"Processing file: {file_name}")

        # Read the CSV file
        df = pd.read_csv(file_name)

        # Group by booking_id and check if there exists at least one entry with booking_status == 'TRIP_ENDED'
        ended_trips = df[df['booking_status'] == 'TRIP_ENDED']
        all_booking_ids = set(df['booking_id'])
        incomplete_booking_ids = set(ended_trips['booking_id'])

        # Filter out booking IDs where there exists an entry with booking_status == 'TRIP_ENDED' while some other entries have a different status
        incomplete_booking_ids = [booking_id for booking_id in all_booking_ids if booking_id not in incomplete_booking_ids]

        # Save incomplete booking IDs to another CSV file
        incomplete_booking_df = pd.DataFrame({'booking_id': incomplete_booking_ids})
        incomplete_booking_df.to_csv(f'incomplete_bookings_{day:02d}.csv', index=False)

        # Total entries count for current file
        total_entries = len(incomplete_booking_ids)
        print(f"Total entries in output CSV file for {file_name}: {total_entries}")
        print("Number of booking IDs:", len(all_booking_ids))
        print("\n")

# Example usage:
process_files("anon_pooling_pings_jan_24_amman_2024-01-", 1, 31)
```
