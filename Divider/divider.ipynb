### Divides a JSON file into 40 parts

```python
import json
import os

def divide_json(json_file_path, output_folder):
    with open(json_file_path, 'r') as f:
        data = []
        for line in f:
            data.append(json.loads(line))
    
    total_records = len(data)
    records_per_file = total_records // 40
    remainder = total_records % 40

    start_index = 0
    for i in range(40):
        end_index = start_index + records_per_file + (1 if i < remainder else 0)
        output_file_path = os.path.join(output_folder, f"part_{i+1}.json")
        with open(output_file_path, 'w') as f:
            for record in data[start_index:end_index]:
                json.dump(record, f)
                f.write('\n')  # Add a newline after each JSON object
        start_index = end_index

# Example usage:
input_json_file = 'original_traj_mapped.json'  # Path to your JSON file
output_folder = 'output'         # Folder where divided files will be saved
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
divide_json(input_json_file, output_folder)
```
